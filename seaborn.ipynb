# Heatmap

```python
flight=sns.load_dataset('flights')
```

```python
flight
```

```python
x=flight.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
sns.heatmap(x)
plt.show()
```

```python
sns.heatmap(x, cbar=False)
plt.show()
```

```python
 sns.heatmap(x, cbar=False, linewidths=0.5)  #distance b/w tw adjacent cells can be increase
plt.show()
```

```python
 sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d')  
plt.show()
```

```python
 sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='winter' )
plt.show()
```

# Clustermap

```python
#type of heatmap but is clustered, cluster rows and columns together on the basis of similarity
#similar rows come together and vice versa
```

```python
 sns.clustermap(x)
plt.show()
```

```python
 sns.clustermap(x, metric='correlation')
plt.show()
```

```python
 sns.clustermap(x,z_score=0, annot=True, metric='correlation')
plt.show()
```

```python
 sns.clustermap(x,z_score=0, annot=True, row_cluster=False, metric='correlation')
plt.show()  #now just on the basis of columns not rows
```

```python
 sns.clustermap(x,z_score=0, annot=True, col_cluster=False, metric='correlation')
plt.show()  #now just on the basis of rows not columns
```

# Jointplot

```python
# multipanel figure that shows both bivariate and univariate on separate axes
#used in 4-manner
#1. scatterplots
#2. hexbin plot
#3. kdeplots
#4. regplot
```

```python
data
```

```python
# 1. Scatterplot
sns.jointplot(x='total_bill', y='tip', data=data)
plt.show() 
#on x-axis histogram of total bill
#on y-axis histrogram of tip
```

```python
# 2. Hexbin plot  #2-D histogram plot
sns.jointplot(x='total_bill', y='tip', kind='hex', data=data)
plt.show()
```

```python
# 2. Kde plot  #2-D pdf
sns.jointplot(x='total_bill', y='tip', kind='kde', data=data)
plt.show()
```

```python
# 2. Regplot --regression  plot, best fit line used for prediction
sns.jointplot(x='total_bill', y='tip', kind='reg', data=data)
plt.show()
```

# Displots

```python
# used to do univariate analysis on conto=inoous/numerical data
# combines 3-things in ine graph
# 1.histogram
# 2. kernel density estimation(PDF)
# 3. Rugplot
```

```python
titanic= sns.load_dataset('titanic')
titanic
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
```

# plotting displot

```python
sns.distplot(titanic['age']) #pdf+hist
plt.show()
```

```python
#plotting only histogram
sns.distplot(titanic['age'], kde=False)
plt.show()
```

```python
sns.distplot(titanic['age'],bins=10, kde=False)
plt.show()
```

```python
#plotting only Kde
sns.distplot(titanic['age'], hist=False)
plt.show()
```

```python
#plotting only rugplot --> on x-axis, side by side
sns.distplot(titanic['age'], hist=False, kde=False, rug=True)
plt.show()
```

```python
#HOW TO USE DISPLOT
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
plt.show()
```
